# Regression Model Representation.

In this file we are learning how the regression model is working. And what is the basic method  to compute regression model in linear regression.


```python
# Import all the important libraries we need to run this file.
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("./deeplearning.mplstyle")
```

## Feature and target values for model.

| Size (1000 sqft)     | Price (1000s of dollars) |
| -------------------| ------------------------ |
| 1.0               | 300                      |
| 2.0               | 500                      |

```python
x_train = np.array([1 ,2.0])
y_train = np.array([300.0 ,500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# Calculate the training examples.
print(f"x_train.shape = {x_train.shape}")
m = x_train.shape[0]
print(f"The number of the training examples is = {m}")
```

```python
# Alternative method to calculate the training examples.
m = len(x_train)
print(f"The number of the training example is ={m}")
```

```python
# Model representation.
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}) ,y^({i})) = ({x_i} , {y_i})")
```

```python
# plot before the model evaluation.

plt.scatter(x_train ,y_train , marker='x',c ='r')
# set title
plt.title('Housing price:')
# Set y-axis 
plt.ylabel('price (in 1000 dollars)')

# set x_axis

plt.xlabel('size (in 1000 sqft)')
plt.show()
```

```python
# Model function.

def compute_model_output(x ,w,b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
w = 100
b = 100
print(f'w = {w}')
print(f'b = {b}')
```

```python
temp_wb = compute_model_output(x_train , w ,b)
plt.plot(x_train ,temp_wb , c ='b' ,label ='our prediction')
plt.scatter(x_train ,y_train , marker='*',c ='b', label = 'actual prediction')
# set titles 
plt.title('Housing price:')
# set y-axis
plt.ylabel('price (in 1000 dollars)')
# set x-axis 
plt.xlabel('size (in 1000sqft)')
plt.legend()
plt.show()
```

```python
# Prediction
w = 200
b = 200
x_i = 1.2
f_wb = w * x_i + b
print(f'Price of of 1200 sqft  = {f_wb}')
```
